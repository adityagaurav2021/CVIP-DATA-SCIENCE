```python
import pandas as pd

# Load the terrorism dataset into a pandas DataFrame
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='ISO-8859-1', low_memory=False)

# Print the first 5 rows of the dataset
print(df.head())

# Print the shape of the dataset
print(df.shape)

# Print the column names of the dataset
print(df.columns)

# Check for missing values in the dataset
print(df.isna().sum())

# Print the descriptive statistics of the dataset
print(df.describe())

# Count the number of unique values in each column
print(df.nunique())

# Count the number of incidents per year
incidents_per_year = df.groupby('iyear').size()
print(incidents_per_year)

# Plot the number of incidents per year
import matplotlib.pyplot as plt
plt.plot(incidents_per_year)
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Global Terrorism Database into a Pandas DataFrame
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='ISO-8859-1', low_memory=False)

# Extract the year from the 'iyear' column and count the number of attacks per year
attacks_per_year = df['iyear'].value_counts().sort_index()

# Create a bar graph of the number of attacks per year
plt.bar(attacks_per_year.index, attacks_per_year.values)

# Set the title and axis labels
plt.title('Number of Terrorist Attacks by Year')
plt.xlabel('iyear')
plt.ylabel('Number of Attacks')

# Display the plot
plt.show()
```
